```python
df_sentences = pd.read_csv('sentences_vf.csv')
print(df_sentences.head(10))
```

```python
#sent2vec needs to be installed from github repository

#import sent2vec

# Load the BioSentVec model
model_path = 'BioSentVec_PubMed_MIMICIII-bigram_d700.bin'
model = sent2vec.Sent2vecModel()
model.load_model(model_path)
print('Model successfully loaded')
```

```python
# Ensure the sentence column is of type string
df_sentences['sentence'] = df_sentences['sentence'].astype(str)
```

```python
# import h5py
# import numpy as np
# import pandas as pd
# import os
# Function to generate embeddings for a single sentence
def get_sentence_embedding(sentence):
    return model.embed_sentence(sentence)

# Function to process data in batches and save to HDF5
def process_and_save_batches(df, batch_size=100, output_dir='processed_embeddings'):
    num_batches = int(np.ceil(len(df) / batch_size))
    
    for i in range(num_batches):
        batch_df = df.iloc[i*batch_size:(i+1)*batch_size]
        batch_embeddings = np.array([get_sentence_embedding(sentence) for sentence in batch_df['sentence']])
        
        # make sure the output directory exists
        os.makedirs(output_dir, exist_ok=True)
        
        # Save the batch to HDF5 file
        file_path = os.path.join(output_dir, f'processed_embeddings_batch_{i+1}.h5')
        with h5py.File(file_path, 'w') as f:
            f.create_dataset('original_index', data=batch_df['original_index'].values)
            f.create_dataset('sentence', data=np.array(batch_df['sentence'].values, dtype=h5py.string_dtype()))
            f.create_dataset('embeddings', data=batch_embeddings)
        
        print(f"Processed and saved batch {i+1}/{num_batches}")


batch_size = 100  
process_and_save_batches(df_sentences, batch_size=batch_size)
```
